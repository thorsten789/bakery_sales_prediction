```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

kiwo = pd.read_csv("https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/kiwo.csv")
umsatzdaten_gekuerzt = pd.read_csv("https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/umsatzdaten_gekuerzt.csv")
wetter = pd.read_csv("https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/wetter.csv")

umsatz_kiwo = umsatzdaten_gekuerzt.merge(kiwo, on='Datum', how='outer')
df = umsatz_kiwo.merge(wetter, on='Datum', how='outer')

df.head()
```

```python
# Generate descriptive statistics for the 'Temperatur' column
temp_stats = df['Temperatur'].describe()
print("Descriptive Statistics for Temperature:")
print(temp_stats)

# Additional useful statistics
print("\nAdditional Statistics:")
print(f"Median temperature: {df['Temperatur'].median()}")
print(f"Variance: {df['Temperatur'].var()}")
print(f"Skewness: {df['Temperatur'].skew()}")
print(f"Kurtosis: {df['Temperatur'].kurt()}")

# Check for missing values
missing_temps = df['Temperatur'].isna().sum()
print(f"\nMissing temperature values: {missing_temps}")
print(f"Percentage of missing values: {(missing_temps / len(df)) * 100:.2f}%")

# Create a histogram to visualize the temperature distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Temperatur'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Create a box plot for temperature
plt.figure(figsize=(8, 6))
plt.boxplot(df['Temperatur'].dropna())
plt.title('Temperature Box Plot')
plt.ylabel('Temperature (°C)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Set a consistent style for all plots
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 12})

# 1. Enhanced histogram with KDE
plt.figure(figsize=(12, 6))
sns.histplot(df['Temperatur'].dropna(), kde=True, bins=25, color='skyblue')
plt.axvline(df['Temperatur'].mean(), color='red', linestyle='--', label=f'Mean: {df["Temperatur"].mean():.2f}°C')
plt.axvline(df['Temperatur'].median(), color='green', linestyle='-.', label=f'Median: {df["Temperatur"].median():.2f}°C')
plt.title('Temperature Distribution with Density Curve')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()

# 2. Box plot by year
# Extract year from date and create a new column
df['Year'] = pd.to_datetime(df['Datum']).dt.year
plt.figure(figsize=(14, 7))
sns.boxplot(x='Year', y='Temperatur', data=df)
plt.title('Temperature Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 3. Time series of temperature
# Create a monthly average temperature dataframe
df_copy = df.copy()
df_copy['Date'] = pd.to_datetime(df_copy['Datum'])
df_copy.set_index('Date', inplace=True)
monthly_temp = df_copy['Temperatur'].resample('M').mean()
plt.figure(figsize=(15, 7))
monthly_temp.plot()
plt.title('Monthly Average Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Seasonal patterns - Monthly averages across all years
df_copy['Month'] = df_copy.index.month
monthly_avg = df_copy.groupby('Month')['Temperatur'].mean()
plt.figure(figsize=(12, 6))
monthly_avg.plot(kind='bar', color='skyblue')
plt.title('Average Temperature by Month (All Years)')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

# 5. Scatter plot: Temperature vs. Umsatz (Sales)
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Temperatur', y='Umsatz', data=df, alpha=0.5, hue='Warengruppe')
plt.title('Temperature vs. Sales by Product Category')
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales (Umsatz)')
plt.grid(True, alpha=0.3)
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()

# 6. Scatter plot with regression line: Temperature vs. Sales (overall)
plt.figure(figsize=(12, 7))
sns.regplot(x='Temperatur', y='Umsatz', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Temperature vs. Sales with Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales (Umsatz)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 7. Temperature heatmap by year and month
# Pivot the data to create a year-month matrix of average temperatures
df_copy['Year'] = df_copy.index.year
temp_pivot = df_copy.pivot_table(index='Year', columns='Month', values='Temperatur', aggfunc='mean')
plt.figure(figsize=(14, 8))
sns.heatmap(temp_pivot, cmap='coolwarm', annot=True, fmt='.1f')
plt.title('Average Temperature Heatmap by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

# 8. Temperature vs. Cloud Cover (Bewoelkung) scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Bewoelkung', y='Temperatur', data=df, alpha=0.5)
plt.title('Temperature vs. Cloud Cover')
plt.xlabel('Cloud Cover')
plt.ylabel('Temperature (°C)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 9. Temperature distribution during Kiel Week vs. non-Kiel Week
plt.figure(figsize=(12, 6))
df['Is_KielerWoche'] = df['KielerWoche'].notna()
sns.boxplot(x='Is_KielerWoche', y='Temperatur', data=df)
plt.title('Temperature Distribution: Kiel Week vs. Regular Days')
plt.xlabel('During Kiel Week')
plt.ylabel('Temperature (°C)')
plt.xticks([0, 1], ['Regular Days', 'Kiel Week'])
plt.tight_layout()
plt.show()
```

```python
# Descriptive statistics for 'Bewoelkung' (Cloud Cover)
cloud_stats = df['Bewoelkung'].describe()
print("Descriptive Statistics for Cloud Cover (Bewoelkung):")
print(cloud_stats)

# Additional statistics
print("\nAdditional Statistics:")
print(f"Median cloud cover: {df['Bewoelkung'].median()}")
print(f"Variance: {df['Bewoelkung'].var()}")
print(f"Skewness: {df['Bewoelkung'].skew()}")
print(f"Kurtosis: {df['Bewoelkung'].kurt()}")

# Check for missing values
missing_cloud = df['Bewoelkung'].isna().sum()
print(f"\nMissing cloud cover values: {missing_cloud}")
print(f"Percentage of missing values: {(missing_cloud / len(df)) * 100:.2f}%")

# Histogram for cloud cover
plt.figure(figsize=(10, 6))
plt.hist(df['Bewoelkung'].dropna(), bins=20, color='lightgray', edgecolor='black')
plt.title('Cloud Cover (Bewoelkung) Distribution')
plt.xlabel('Cloud Cover')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# 1. Time series plot of daily cloud cover
df['Datum'] = pd.to_datetime(df['Datum'])
plt.figure(figsize=(15, 6))
plt.plot(df['Datum'], df['Bewoelkung'], alpha=0.5)
plt.title('Daily Cloud Cover (Bewoelkung) Over Time')
plt.xlabel('Date')
plt.ylabel('Cloud Cover')
plt.tight_layout()
plt.show()

# 2. Monthly average cloud cover over time
df['YearMonth'] = df['Datum'].dt.to_period('M')
monthly_cloud = df.groupby('YearMonth')['Bewoelkung'].mean().reset_index()
monthly_cloud['YearMonth'] = monthly_cloud['YearMonth'].dt.to_timestamp()
plt.figure(figsize=(15, 6))
plt.plot(monthly_cloud['YearMonth'], monthly_cloud['Bewoelkung'], marker='o')
plt.title('Monthly Average Cloud Cover Over Time')
plt.xlabel('Month')
plt.ylabel('Average Cloud Cover')
plt.tight_layout()
plt.show()

# 3. Seasonal pattern: Average cloud cover by month (across all years)
df['Month'] = df['Datum'].dt.month
monthly_avg_cloud = df.groupby('Month')['Bewoelkung'].mean()
plt.figure(figsize=(10, 5))
monthly_avg_cloud.plot(kind='bar', color='gray')
plt.title('Average Cloud Cover by Month (All Years)')
plt.xlabel('Month')
plt.ylabel('Average Cloud Cover')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
```

```python
# Create bins for cloud cover
df['Bewoelkung_bin'] = pd.cut(df['Bewoelkung'], bins=5)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Bewoelkung_bin', y='Umsatz', data=df)
plt.title('Sales (Umsatz) by Cloud Cover (Bewoelkung) Bins')
plt.xlabel('Cloud Cover (Binned)')
plt.ylabel('Sales (Umsatz)')
plt.tight_layout()
plt.show()
mean_sales_by_cloud = df.groupby('Bewoelkung')['Umsatz'].mean()
plt.figure(figsize=(10, 6))
mean_sales_by_cloud.plot()
plt.title('Average Sales (Umsatz) by Cloud Cover (Bewoelkung)')
plt.xlabel('Cloud Cover')
plt.ylabel('Average Sales (Umsatz)')
plt.tight_layout()
plt.show()
# Example: Average sales by month and cloud cover bin
df['Month'] = df['Datum'].dt.month
pivot = df.pivot_table(index='Month', columns='Bewoelkung_bin', values='Umsatz', aggfunc='mean')
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Average Sales by Month and Cloud Cover Bin')
plt.xlabel('Cloud Cover (Binned)')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

```python
# Ensure 'Bewoelkung_bin' and 'YearMonth' columns exist
df['Bewoelkung_bin'] = pd.cut(df['Bewoelkung'], bins=5)
df['YearMonth'] = df['Datum'].dt.to_period('M')

# Pivot table: rows=Month, columns=Cloud Cover Bin, values=mean sales
pivot = df.pivot_table(index='YearMonth', columns='Bewoelkung_bin', values='Umsatz', aggfunc='mean')

plt.figure(figsize=(14, 7))
pivot.plot(ax=plt.gca())
plt.title('Monthly Average Sales by Cloud Cover Bin Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Average Sales (Umsatz)')
plt.legend(title='Cloud Cover (Binned)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
