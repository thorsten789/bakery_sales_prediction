The CSV file was generated by claude.ai with data for school holidays for Schleswig-Holstein which was taken by manual copy-and-paste on 22.05.2025 from this page:

https://www.kalenderpedia.de/ferien/ferien-schleswig-holstein-2019.html

As code numbers for the weekdays, Claude was told to have Mondays as 1 and Sundays as 7. But checked against the calendar, it turned out that way:

* 1 - Sunday (for example 01.01.2012)
* 2 - Monday
* 3 - Tuesday (for example 31.12.2019)
* 4 - Wednesday
* 5 - Thursday
* 6 - Friday
* 7 - Saturday
